```python
import numpy as np
import matplotlib.pyplot as plt
import gym
from gym import error, spaces, utils
from gym.utils import seeding
from enum import Enum

class Plant:
    def __init__(self, species, maturity=110):
        self.species = species
        self.maturity = maturity
        self.age = 0
        
    def __repr__(self):
        return "{}".format(self.species)
    

class Field(gym.Env):
    metadata = {'render.modes': ['human']}

    def __init__(self, size=5, sow_limit=200, season=120, calendar=0):
        # parameters for overall field character
        self.size = size
        self.sow_limit = sow_limit
        self.season = season
        self.calendar = calendar
        
        # constants for computing end-of-season reward---distances represent meters
        self.crowding_dist = .02
        self.maize_maize_dist = .1
        self.bean_support_dist = .1
        self.crowding_penalty = .1
        self.maize_maize_penalty = .9
        self.bean_support_bonus = .6
        
        # OpenAI action and observation space specifications
        self.action_space = spaces.Discrete(4)
        # self.observation_space = spaces.???
        
        # field is initialized by calling reset()
        self.field = None
        
    def step(self, action, sow_limit=1):
        # first, sow a plant (or wait) depending on action chosen
        # could be cleaned up with plants as an enumeration?
        i = 0
        while i < sow_limit:
            if action == 0:
                self.field = np.append(self.field, [[self.size * np.random.random(), 
                                             self.size * np.random.random(), 
                                             Plant('Maize')]], axis=0)
            elif action == 1:
                self.field = np.append(self.field, [[self.size * np.random.random(), 
                                             self.size * np.random.random(), 
                                             Plant('Bean')]], axis=0)
            elif action == 2:
                self.field = np.append(self.field, [[self.size * np.random.random(), 
                                             self.size * np.random.random(), 
                                             Plant('Squash')]], axis=0)
            # when action == 3, nothing is done (agent waits)   
            i += 1
        
        # increment timekeeping
        self.calendar +=1
        for plant in self.field:
            plant[2].age += 1
            
        done = self.calendar == self.season
            
        if not done:
            reward = 0
        else:
            reward = self.get_reward()
            
        return self.field, reward, done, {}
    
    def reset(self):
        # field is initialized with one random corn plant in order to make sowing (by np.append) work
        self.field = np.array([[self.size * np.random.random(), 
                                self.size * np.random.random(), 
                                Plant('Maize')]])
        
    def render(self, mode='human'):
        # derived arrays for each plant type so that they can be plotted separately
        # include else clause in list comprehensions to prevent the plot breaking?
        is_maize = np.array([row for row in self.field if row[2].__repr__() == 'Maize'])
        is_bean = np.array([row for row in self.field if row[2].__repr__() == 'Bean'])
        is_squash = np.array([row for row in self.field if row[2].__repr__() == 'Squash'])
        
        # plot the field---currently breaks if any plant type is absent
        plt.figure(figsize=(10, 10))
        plt.scatter(is_maize[:,0], is_maize[:,1], c='green', s=200, marker = 'o', alpha=.5, edgecolor='#303030')
        plt.scatter(is_bean[:,0], is_bean[:,1], c='brown', s=150, marker = 'o', alpha=.5, edgecolor='#303030')
        plt.scatter(is_squash[:,0], is_squash[:,1], c='orange', s=400, marker = 'o', alpha=.5, edgecolor='#303030')

        plt.show()
        
        print("\n---\nTotal yield in Calories is {}.\n---\n".format(round(reward, 1)))
    
    def close(self):
        # unneeded right now? AFAICT this is only used to shut down realtime movie visualizations
        pass
    
    def get_reward(self):
        # array of plant coordinates for computing distances
        xy_array = np.array([[row[0], row[1]] for row in self.field])

        # distances[m,n] is distance from mth to nth plant in field
        distances = np.linalg.norm(xy_array - xy_array[:,None], axis=-1)
        
        reward = 0
        i = 0
        while i < len(self.field):
            if self.field[i,2].age < self.field[i,2].maturity:
                reward += 0
            elif self.field[i,2].__repr__() == 'Maize':
                cal = 1
                j = 0
                while j < len(distances[0]):
                    if (self.field[j,2].__repr__() == 'Bean' 
                            and distances[i,j] < self.bean_support_dist):
                        cal += self.bean_support_bonus
                    if (self.field[j,2].__repr__() == 'Maize' 
                            and i !=j 
                            and distances[i,j] < self.maize_maize_dist):
                        cal *= self.maize_maize_penalty
                    if 0 < distances[i,j] < self.crowding_dist:
                        cal *= self.crowding_penalty
                    j += 1
                reward += cal
            elif self.field[i,2].__repr__() == 'Bean':
                reward += .25
            elif self.field[i,2].__repr__() == 'Squash':
                reward += 3
            i += 1        
        return reward
```

### Third Attempt at floating-point sowing and rewards

I now have the beginnings of an object-oriented approach.  Sowing is done as a method of the Field class, and can sow one specified plant at a time.  Stepping time forward is also possible as a Field method—however, at the moment all plants are sowed on day 1, so they all reach maturity and thus time doesn't matter for yield.

The code to calculate end-of-episode reward now handles a generic crowding penalty, maize-bean support, and maize-maize penalty.  The way that it checks for plants of a certain type within a certain distance is, I think, sufficient to extend to the rest of the plant interactions that were outlined back in the google spreadsheet.  (That table of interactions is included at the bottom of this notebook.)  Obviously this is extremely simplistic relative to real biological interactions, but it's a start.

I am hoping that this model, when finished, can both serve as the testbed for the agent bake-off paper and also yield the illustrative result for the beyond-the-row paper.

```python
import numpy as np
import matplotlib.pyplot as plt

class Plant:
    def __init__(self, species, maturity=110):
        self.species = species
        self.maturity = maturity
        self.age = 0
        
    def __repr__(self):
        return "{}".format(self.species)
    

class Field:
    def __init__(self, size=5, sow_limit=200, season=120):
        # field is initialized with one random corn plant in order to make later code work
        self.size = size
        self.sow_limit = sow_limit
        self.season = season
        self.field = np.array([[self.size * np.random.random(), 
                                self.size * np.random.random(), 
                                Plant('Maize')]])
        
    def sow_maize(self):
        self.field = np.append(self.field, [[self.size * np.random.random(), 
                                             self.size * np.random.random(), 
                                             Plant('Maize')]], axis=0)
        
    def sow_bean(self):
        self.field = np.append(self.field, [[self.size * np.random.random(), 
                                             self.size * np.random.random(), 
                                             Plant('Bean')]], axis=0)
        
    def sow_squash(self):
        self.field = np.append(self.field, [[self.size * np.random.random(), 
                                             self.size * np.random.random(), 
                                             Plant('Squash')]], axis=0)
        
    def time_step(self):
        for plant in self.field:
            plant[2].age += 1
        

# create field, sow randomly, advance time to end of season
testbed = Field()
actions = [testbed.sow_maize, testbed.sow_bean, testbed.sow_squash]
for i in range(testbed.sow_limit):
    np.random.choice(actions)()
for i in range(testbed.season):
    testbed.time_step()
    
# derived arrays for each plant type so that they can be plotted separately
# include else clause in list comprehensions to prevent the plot breaking?
is_maize = np.array([row for row in testbed.field if row[2].__repr__() == 'Maize'])
is_bean = np.array([row for row in testbed.field if row[2].__repr__() == 'Bean'])
is_squash = np.array([row for row in testbed.field if row[2].__repr__() == 'Squash'])

# array of plant coordinates for computing distances
xy_array = np.array([[row[0], row[1]] for row in testbed.field])

# array of each plant's distance from each other plant
# distances[m,n] is distance from mth to nth plant in field
distances = np.linalg.norm(xy_array - xy_array[:,None], axis=-1)
    
    
# constants for computing reward---distance values represent meters
# make these properties of the field/plant, à la days to maturity?
total_reward = 0

crowding_dist = .02
maize_maize_dist = .1
bean_support_dist = .1

crowding_penalty = .1
maize_maize_penalty = .9

bean_support_bonus = .6


# computes overall reward (yield) given plant locations and types
def get_reward(reward):
    '''
    - iterate through plants, checking first for maturity
    - given each plant species, adjusts reward based on neighbors
    - feels very cumbersome---is there a better way to do this?
    '''
    i = 0
    while i < len(testbed.field):
        if testbed.field[i,2].age < testbed.field[i,2].maturity:
            reward = 0
        elif testbed.field[i,2].__repr__() == 'Maize':
            cal = 1
            j = 0
            while j < len(distances[0]):
                if (testbed.field[j,2].__repr__() == 'Bean' 
                        and distances[i,j] < bean_support_dist):
                    cal += bean_support_bonus
                if (testbed.field[j,2].__repr__() == 'Maize' 
                        and i !=j 
                        and distances[i,j] < maize_maize_dist):
                    cal *= maize_maize_penalty
                if 0 < distances[i,j] < crowding_dist:
                    cal *= crowding_penalty
                j += 1
            reward += cal
        elif testbed.field[i,2].__repr__() == 'Bean':
            reward += .25
        elif testbed.field[i,2].__repr__() == 'Squash':
            reward += 3
        i += 1        
    print("\n---\nTotal yield in Calories is {}.\n---\n".format(round(reward, 1)))
    
    
    

# compute yield
get_reward(total_reward)

# plot the field---currently breaks if any plant type is absent
plt.figure(figsize=(10, 10))
plt.scatter(is_maize[:,0], is_maize[:,1], c='green', s=200, marker = 'o', alpha=.5, edgecolor='#303030')
plt.scatter(is_bean[:,0], is_bean[:,1], c='brown', s=150, marker = 'o', alpha=.5, edgecolor='#303030')
plt.scatter(is_squash[:,0], is_squash[:,1], c='orange', s=400, marker = 'o', alpha=.5, edgecolor='#303030')

plt.show()
```
